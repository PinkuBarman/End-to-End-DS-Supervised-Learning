```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. In Iris Dataset, plot a scatter plot differentiating each classifier with a boundary.
```

```python
Iris_df = pd.read_csv(r'E:\End to End DS - Supervised Learning\Data Source\iris.csv')
```

```python
Iris_df.sample(5)
```

```python
Iris_df.plot(kind = 'scatter',x ='Sepal.Length',y ='Petal.Length')
plt.title('Diffentiate Sepal Vs Petal Length')
plt.show()
```

```python
Iris_df.plot(kind = 'scatter',x ='Sepal.Width',y ='Petal.Width')
plt.title('Diffentiate Sepal Vs Petal Width')
plt.show()
```

```python
# Plot using Seaborn

# style used as a theme of graph  
# for example if we want black  
# graph with grid then write "blackgrid" 
sns.set_style('whitegrid')


# sepal_length, petal_length are iris 
# feature data height used to define 
# Height of graph whereas hue store the 
# class of iris dataset. 
sns.FacetGrid(Iris_df,hue= "Species", height = 6).map(plt.scatter, 'Sepal.Length','Petal.Length').add_legend()

plt.show()
```

```python
# 2. In wine quality dataset, plot a correlation matrix for all numerical variables and find out which columns are 
# highly correlated to each other.
```

```python
# winequality-red

winequality_df = pd.read_csv(r'E:\End to End DS - Supervised Learning\Data Source\winequality-red.csv')

plt.figure(figsize = (10,10))
corrmatrix = winequality_df.corr()
sns.heatmap(corrmatrix,color = "k", annot=True)
plt.show()

winequality_df.sample(5)
```

```python
# 1. Density and Fixed Acidity are highly correlated
# 2. Free sulphul dioxide and total sulphur dioxide are highly correlated
# 3. Cirix acid and fixed acidity are highly correlated
```

```python
# seaborn.catplot  show the relationship between a numerical and one or more categorical variables

sns.catplot(x = 'quality', data = winequality_df , kind = 'count')
plt.show()

# quality has a high number of values in categories 5, 6 and 7.
# Only a few observations are there for the categories 3 & 8.
```
